### Ideas

- **`if __name__ == "__main__"`** — code that runs **only** when you execute the file directly, not when it is imported.

## Part 1 — Read environment variables safely

`os.environ` is like a dict of string settings. Use **`.get("NAME", default)`** when a variable is optional.

Run the cell (it uses defaults if unset).

````python
import os

log_level = os.environ.get("LOG_LEVEL", "INFO")
batch_size_raw = os.environ.get("BATCH_SIZE", "100")

print("LOG_LEVEL:", log_level)
print("BATCH_SIZE raw:", batch_size_raw)
````

## Part 2 — Parse a string to int safely

If `BATCH_SIZE` is not a valid integer, fall back to `100` and **log** a clear error.

````python
import logging
import os

logging.basicConfig(level=logging.INFO)
log = logging.getLogger("topic5")

raw = os.environ.get("BATCH_SIZE", "100")
try:
    batch_size = int(raw)
except ValueError:
    log.error("bad BATCH_SIZE=%r — using 100", raw)
    batch_size = 100

print("batch_size:", batch_size)
````

## Part 3 — Entry point pattern

Small scripts often use:

```python
def main():
    return 0  # success

if __name__ == "__main__":
    raise SystemExit(main())
```

**Why:** importing the file (for tests) should not **auto-run** the job; running `python script.py` should.

````python
def main():
    print("main ran")
    return 0

if __name__ == "__main__":
    raise SystemExit(main())
````
