Files are how you first touch real data: exports from tools, uploads from vendors, dumps from APIs saved as `.csv` or `.json`.

### New ideas in this notebook

- **Path** — a safe way to talk about file locations (folders and file names).
- **`with open(...)`** — opens a file and **closes** it automatically when the block ends.
- **`csv.DictReader`** — reads a CSV so each row becomes a **dict** (column name → cell text).
- **`json.dump`** — writes Python lists/dicts as JSON text.

**Paths:** this notebook lives in `week1/notebooks/`. Sample data is at `../data/sample/vendor_a.csv` (one folder up, then `data`, then `sample`).

## Part 1 — Paths with `pathlib`

`Path("folder") / "file.csv"` joins parts using the correct slashes for your OS.

Run the cell (change the path if your project layout differs).

```python
from pathlib import Path

here = Path.cwd()
sample = Path("..") / "data" / "sample" / "vendor_a.csv"
print("current dir:", here)
print("sample csv:", sample.resolve())
print("exists:", sample.resolve().exists())
```

## Part 2 — Open a text file read-only

Mode `"r"` means read. Mode `"w"` means write (careful: can overwrite). `"rt"` is read text.

Using **`with`** means: even if an error happens, Python closes the file for you.

```python
from pathlib import Path

p = Path("..") / "data" / "sample" / "vendor_a.csv"
with p.open("r", encoding="utf-8") as f:
    first_line = f.readline()
print(repr(first_line))
```

## Part 3 — Read CSV rows as dicts

`csv.DictReader` uses the **first row as headers**. Each later row becomes `{"ProductID": "...", ...}`.

Run the cell — it prints how many data rows were read.

```python
import csv
from pathlib import Path

p = Path("..") / "data" / "sample" / "vendor_a.csv"
with p.open(newline="", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    fieldnames = reader.fieldnames
    rows = list(reader)
print("columns:", fieldnames)
print("number of rows:", len(rows))
print("first row:", rows[0] if rows else None)
```

## Part 4 — Write JSON

`json.dump(obj, file, indent=2, ensure_ascii=False)` writes readable JSON. `ensure_ascii=False` keeps non-English characters readable.

We often create an **output folder** first with `.mkdir(parents=True, exist_ok=True)`.

```python
import json
from pathlib import Path

out_dir = Path("..") / "data" / "output"
out_dir.mkdir(parents=True, exist_ok=True)
out_file = out_dir / "topic3_demo.json"

sample_obj = {"hello": "world", "rows": 3}
with out_file.open("w", encoding="utf-8") as f:
    json.dump(sample_obj, f, indent=2, ensure_ascii=False)

print("wrote", out_file.resolve())
```
