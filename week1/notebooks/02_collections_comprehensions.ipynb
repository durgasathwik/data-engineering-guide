- **List** — ordered items (keep order of arrival).
- **Set** — unique items (fast "have I seen this?").
- **Dict** — map a **key** to a **value** (lookup tables, counts).

## Part 1 — List: ordered sequence

A **list** uses square brackets: `[1, 2, 3]`. Order matters. **`append`** adds one item to the end.

Run the example.

````python
seen_order = []
seen_order.append("first")
seen_order.append("second")
seen_order
````

## Part 2 — Set: unique membership

A **set** only stores each value **once**. Use **`x in my_set`** to test membership.

Run the example.

````python
unique = set()
unique.add("a")
unique.add("a")
unique.add("b")
print(unique)
print("a" in unique)
````

## Part 3 — Dict as lookup table

**`dict.get(key, default)`** returns `default` if `key` is missing (no crash).

Run the example.

````python
code_to_name = {"EL": "Electronics", "HM": "Home"}
print(code_to_name["EL"])
print(code_to_name.get("XX", "UNKNOWN"))
````

## Part 4 — Tuples `(a, b)`

A **tuple** is a short fixed group in order, e.g. `(product_id, vendor_id)`.

Run the example.

````python
pair = ("SKU-1", "vendor_A")
print(pair[0], pair[1])
````

## Part 5 — List comprehension (read-only intro)

A compact way to build a list from another list:

```python
[n * 2 for n in numbers]
```

Read: "for each n in numbers, put n*2 in the new list." Challenge C can use a normal loop instead if you prefer.

````python
numbers = [1, 2, 3]
doubled = [n * 2 for n in numbers]
doubled
````
