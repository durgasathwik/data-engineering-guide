# Week 1 — Topic 4: When things go wrong (errors and logs)

**Pipelines fail** in the real world: a file is missing, one row is corrupted, a number is not a number. Good code **decides what to do** instead of stopping blindly—or stops **on purpose** with a clear message.

### Vocabulary

- **Exception** — Python’s way to signal “something went wrong.”
- **`try` / `except`** — “try this; if a specific error happens, handle it.”
- **Logging** — structured messages you can search in production (better than many `print` calls).

## Part 1 — Configure logging once

Run this cell first in the notebook session.

```python
import logging

logging.basicConfig(level=logging.INFO, format="%(levelname)s %(name)s: %(message)s")
log = logging.getLogger("week1.topic4")
log.info("logging is ready")
```

## Part 2 — Catch a specific error

Opening a missing file raises **`FileNotFoundError`**. We can catch **only** that type so other bugs still show up.

Read the code — then temporarily change the path to a file that exists on your machine.

```python
from pathlib import Path

p = Path("/path/that/does/not/exist.txt")
try:
    text = p.read_text(encoding="utf-8")
except FileNotFoundError:
    text = None
    print("No file — handled")

print("text is", text)
```

## Part 3 — Log a warning instead of crashing

When a file is optional, **`log.warning(...)`** records the issue without stopping the whole job.

```python
import logging
from pathlib import Path

log = logging.getLogger("week1.topic4.demo")
p = Path("not_there_xyz.txt")
if not p.exists():
    log.warning("missing file: %s", p)
```
